```python
import pandas as pd
import numpy as np
```

Importing the Bureau Dataset

```python
bureau = pd.read_csv('bureau_score.csv')
bureau
```

Selecting the variables we need into a new dataframe(debt_history)

```python
debt_history = bureau[['CUSTOMER_UNIQUE_ID', 'BAL_30DPD', 'BAL_60DPD', 'BAL_PL_90DPD']]
```

Performing the the SPLIT-APPLY-COMBINE action GROUP_BY function) to calculate the total debt history per customer.

```python
debt_history = debt_history.groupby(['CUSTOMER_UNIQUE_ID']).sum
#performing the sum function on the BAL_30DPD, BAL_60DPD and the BAL_90DPD 
```
